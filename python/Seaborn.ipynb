```python
sns.relplot(x="total_bill",y="tip",hue="time",style="smoker",data=ds)

# sns.relplot(x=ds["total_bill"],y=ds["tip"],hue=ds["time"],style=ds["smoker"])
# both are same
```

```python
sns.relplot(x="total_bill",y="tip",hue="time",style="smoker",data=ds,kind="scatter",col="smoker")
```

```python
sns.relplot(x="total_bill",y="tip",hue="time",style="smoker",data=ds,kind="scatter",row="time")
```

```python
ds["day"].unique()
```

```python
sns.relplot(x="total_bill",y="tip",hue="day",data=ds)
```

```python
sns.relplot(x="total_bill",y="tip",hue="day",data=ds,kind="scatter",col="time")
```

```python
sns.relplot(x="total_bill",y="tip",data=ds)
plt.xlabel("total_bill",color="blue",size=20)
plt.ylabel("tip",color="black",size=20)
plt.title("total bill vs tips")
```
