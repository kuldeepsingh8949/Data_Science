# *Matplotlib*

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x_point = np.array([0,8])
y_point = np.array([0,150])
```

```python
y_point
```

```python
plt.plot(x_point,y_point)  ## py plot
plt.show()
```

```python
x_point = np.array([1,30])
y_point = np.array([5,150])
plt.plot(x_point,y_point,'o')   ## show starting point and end point
plt.show()
```

```python
x_point = np.array([2,10])
plt.plot(x_point,marker='o')   ## marker draw the line between 2 point
plt.show()
```

```python
x_axis = np.array([2,60,2,15,96])
plt.plot(x_axis,marker='o')     ## with multiple point
plt.show()
```

# *space plot*

```python
x_axis = np.array([2,60,2,15,96])
plt.plot(x_axis,marker='*')     ## for star
plt.show()
```

```python
x_point = np.array([1,4,7,5])
plt.plot(x_point,'o:b')       ## for dot line and bold point
plt.show()
```

```python
x_point = np.array([2,3,6,8])
plt.plot(x_point,marker='*',ms=25,mec='r')## mec = boundry colouring
plt.show()                                ## ms = size of point
```

```python
x_axis = np.array([2,6,9,10])
plt.plot(x_axis,marker='*',ms=26,mfc='b')   ## full point colouring
plt.show()
```

```python
x_axis = np.array([2,4,6,8])
plt.plot(x_axis,ls='dotted')  ## ls means full line change like this dotted
plt.show()
```

```python
x_axis = np.array([2,3,4,5])
plt.plot(x_axis,linewidth=11)    ## line and width size increase
plt.show()
```

```python
x_axis = np.array([2,8,18,27])
y_axis = np.array([3,4,7,9])
plt.plot(x_axis,marker='*',ms=26,mfc='b')
plt.plot(y_axis,marker='*',ms=26,mfc='r')
plt.show()
```

```python
x1 = np.array([2,8,18,91])
y1 = np.array([2,37,82,43])
x2 = np.array([2,3,6,8])
y2 = np.array([72,27,29,75])
plt.plot(x1)
plt.plot(x2)
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
x1 = np.array([2,3,5,7])
y1 = np.array([5,8,4])
plt.xlabel("x1")
plt.ylabel("y1")
plt.title("ploting")
plt.plot(x1)
plt.plot(y1)
plt.show()
```

```python
x=[2,3,4,5]
y=[9,8,7,6]
x_axis=np.array(x)
y_axis=np.array(y)
plt.plot(x_axis,y_axis)
plt.grid()
plt.show()
```

```python
x=[2,3,4,5]
y=[9,8,7,6]
x_axis=np.array(x)
y_axis=np.array(y)
plt.subplot(1,2,1)
plt.plot(x_axis,y_axis)
x=[9,8,7,6]
y=[2,3,4,5]
x_axis=np.array(x)
y_axis=np.array(y)
plt.subplot(1,2,2)
plt.plot(x_axis,y_axis)
plt.show()

```

# *Matplotlib Scatter*

```python
x=np.array([2,3,4,5,3,8,7,5,8,5])
y=np.array([2,3,4,5,6,7,8,9,3,4])
size = np.array([1,2,3,4,5,6,2,5,7,9])
plt.scatter(x,y,s=size,alpha=0.5)
plt.show()
```

```python
x=np.array([2,3,4,5,3,8,7,5,8,5])
y=np.array([2,3,4,5,6,7,8,9,3,4])
plt.scatter(x,y)
plt.show()
```

```python
x=np.array([2,3,4,5,3])
y=np.array([2,3,4,5,6])
colour = np.array(['red','blue','green','black','yellow'])
plt.scatter(x,y,c=colour)
plt.show()
```

```python
x=np.array([2,3,4,5,3])
y=np.array([2,3,4,5,6])
size=np.array([20,3,5,70,8])
plt.scatter(x,y,s=size)
plt.show()
```

```python
x=np.array([2,3,4,5,7])
y=np.array([1,3,5,7,9])
size=np.array([20,30,50,70,80])
colour = np.array(['red','blue','green','black','yellow'])
plt.scatter(x,y,s=size,c=colour,alpha=0.5)
plt.show()
```

```python
x=np.array([2,3,4,5,7])
y=np.array([1,3,5,7,9])
size=np.array([20,30,50,70,80])
colour = np.array(['red','blue','green','black','yellow'])
plt.scatter(x,y,s=size,c=colour,cmap='nipy_spectral')
plt.colorbar()
plt.show()
```

# *Bar Plot*

```python
x=np.array([2,3,4,5,7])
y=np.array([1,3,5,7,9])
plt.bar(x,y)
plt.show()
```

```python
x=np.array([2,3,4,5,7])
y=np.array([1,3,5,7,9])
plt.barh(x,y)
plt.show()
```

```python
x=np.array([2,3,4,5,7])
y=np.array([1,3,5,7,9])
plt.barh(x,y,color='black')
plt.show()
```

```python
x=np.array([2,3,4,5,7])
y=np.array([1,3,5,7,9])
plt.bar(x,y,color='black',width=0.2)
plt.show()
```

# *Piechart*

```python
x = np.array([1,2,3,4])
plt.pie(x)
plt.show()
```

```python
x = np.array([1,2,3,4])
label = ["orange","mango","banana","grapes"]
plt.pie(x,labels=label)
plt.show()
```

```python
x = np.array([1,2,3,4])
label = ["orange","mango","banana","grapes"]
plt.pie(x,labels=label,startangle=90)
plt.show()
```

```python
x = np.array([1,2,3,4])
label = ["orange","mango","banana","grapes"]
myexplode = [0.2,0.1,0,0.1]
plt.pie(x,labels=label,startangle=90,explode=myexplode)
plt.show()
```

```python
x = np.array([1,2,3,4])
label = ["orange","mango","banana","grapes"]
myexplode = [0.2,0.1,0,0.1]
plt.pie(x,labels=label,startangle=90,explode=myexplode,shadow=True)
plt.show()
```

```python
x=np.array([1,2,4,5])
mylabel=["orange","mango","banana","apple"]
myexplode=[0.2,0.1,0,0.2]
mycolor=["red","green","blue","yellow"]
plt.pie(x,labels=mylabel,startangle=90,explode=myexplode,shadow=True,colors=mycolor,autopct="%1.2f%%")
plt.legend(title="color")
plt.show()
```
