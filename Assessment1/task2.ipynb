```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Copy of Screenshot 2025-03-28 072452(1).csv")
```

```python
df
```

```python
plt.figure(figsize=(8, 5))
df['Department\nHR'].value_counts().plot(kind='bar', color=['blue', 'orange', 'green'])
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.title("Number of Employees per Department")
plt.show()
```

```python
high_salary_df = df[df['Salary\n50000'] > 60000]
print("Employees with salary greater than 60,000:")
print(high_salary_df)
```

```python
plt.figure(figsize=(8, 5))
for dept in df['Department\nHR'].unique():
    dept_data = df[df['Department\nHR'] == dept]
    plt.plot(dept_data['Name\nAlice'], dept_data['Salary\n50000'], marker='o', linestyle='-', label=dept)

plt.xlabel("Employee ID")
plt.ylabel("Salary")
plt.title("Salary Trend by Department")
plt.legend()
plt.show()
```
